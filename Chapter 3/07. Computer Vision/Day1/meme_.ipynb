```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('meme.jpg')
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
meme_image = rgb_img.copy()

cv2.putText(meme_image,"When you've just realised that,",(160,65),cv2.FONT_HERSHEY_COMPLEX,2.5,(0,0,0),12)
cv2.putText(meme_image,"When you've just realised that,",(160,65),cv2.FONT_HERSHEY_COMPLEX,2.5,(255,255,255),8)

cv2.putText(meme_image,'tomorrow is a holiday',(210,870),cv2.FONT_HERSHEY_COMPLEX,3,(0,0,0),12)
cv2.putText(meme_image,'tomorrow is a holiday',(210,870),cv2.FONT_HERSHEY_COMPLEX,3,(255,255,255),8)

plt.figure(figsize=(14,10))
plt.imshow(meme_image)
```
