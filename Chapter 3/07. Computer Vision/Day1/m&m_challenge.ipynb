```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
colormap = cv2.imread('img/hsv.png')
colormap_rgb = cv2.cvtColor(colormap, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.imshow(colormap_rgb)
```

```python
bgr_mnms = cv2.imread('img/mnms.png')
rgb_mnms = cv2.cvtColor(bgr_mnms, cv2.COLOR_BGR2RGB)
hsv_mnms = cv2.cvtColor(bgr_mnms, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(10,14))

plt.imshow(rgb_mnms)
```

```python
lower_bounds_blue = (95,15,0)
upper_bounds_blue = (135,255,255)
mask_blue = cv2.inRange(hsv_mnms, lower_bounds_blue, upper_bounds_blue)

lower_bounds_green = (40,0,0)
upper_bound_green = (75,255,255)
mask_green = cv2.inRange(hsv_mnms, lower_bounds_green, upper_bound_green)

lower_bounds_red = (12,0,0)
upper_bounds_red = (160,255,255)
mask_red = cv2.inRange(hsv_mnms, lower_bounds_red, upper_bounds_red)


mnms = bgr_mnms.copy()
# mnms[(mask_blue==0)] = [0,0,0]
# mnms[(mask_green==0)] = [0,0,0]
# mnms[(mask_red!=0)] = [0,0,0]
# mnms[(mask_blue==0) & (mask_green==0)] = [0,0,0]
mnms[(mask_blue==0) & (mask_green==0) & (mask_red!=0)] = [0,0,0]



plt.figure(figsize=(10,14))
plt.imshow(cv2.cvtColor(mnms, cv2.COLOR_BGR2RGB))
```

```python
mnms = bgr_mnms.copy()
mnms_f = cv2.cvtColor(mnms, cv2.COLOR_BGR2RGBA)
final_mask = (mask_blue==0) & (mask_green==0) & (mask_red!=0)

mnms_f[final_mask]=mnms_f[final_mask]*0

plt.figure(figsize=(10,14))
plt.imshow(mnms_f)
```
