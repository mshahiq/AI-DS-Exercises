```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('test.jpg')

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_copy = rgb_img.copy()
cv2.rectangle(img_copy, (390,200),(585,360),(255,255,255),2)
cv2.putText(img_copy, 'Tram', (390,190), cv2.FONT_HERSHEY_PLAIN, 2, (0,255,255),3)

cv2.rectangle(img_copy, (190,240),(270,340),(255,255,255),2)
cv2.putText(img_copy, 'Tram', (190,230), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255),3)

cv2.rectangle(img_copy, (595,130),(720,360),(0,255,0),2)
cv2.putText(img_copy, 'Tree', (590,130), cv2.FONT_HERSHEY_PLAIN, 2, (0,255,0),3)

cv2.rectangle(img_copy, (5,5),(185,400),(0,255,0),2)
cv2.putText(img_copy, 'Tree', (185,50), cv2.FONT_HERSHEY_PLAIN, 2, (0,255,0),3)


plt.imshow(img_copy)
plt.show()
```
