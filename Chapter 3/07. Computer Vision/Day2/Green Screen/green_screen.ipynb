Imports

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

imshow() Function

```python
def imshow(raw_pic):
    modified = cv2.cvtColor(raw_pic,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(12,14))
    return plt.imshow(modified)
```

Colormap

```python
colormap_rgb = cv2.imread('img/hsv.png')
imshow(colormap_rgb)
```

Reading images

```python
background = cv2.imread('berlin1920.jpg')
robot = cv2.imread('1920bb8-yellow.jpg')

background_rgb = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
robot_rgb = cv2.cvtColor(robot, cv2.COLOR_BGR2RGB)
```

Converting the images to HSV

```python
background_hsv = cv2.cvtColor(background, cv2.COLOR_BGR2HSV)
robot_hsv = cv2.cvtColor(robot, cv2.COLOR_BGR2HSV)
```

```python
robot_hsv_copy = robot_hsv.copy()
```

Selecting color from ColorMap

```python
lower_bounds = (10, 60, 0)
upper_bounds = (50, 255, 255)
```

Mask Region

```python
mask = cv2.inRange(robot_hsv, lower_bounds, upper_bounds)

masked_background = background_rgb.copy()
masked_robot = robot_rgb.copy()
```

Applying Mask

```python
masked_robot[mask != 0] = [0,0,0]
masked_background[mask == 0] = [0,0,0]
merged = masked_robot + masked_background
final_image = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)
```

Display the images


```python
imshow(robot)
```

```python
imshow(background)
```

```python
imshow(final_image)
```
