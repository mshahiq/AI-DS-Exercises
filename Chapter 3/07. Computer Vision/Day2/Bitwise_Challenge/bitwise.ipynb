### Use AND, OR, NOT, XOR (logic gates)

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
background = np.zeros([800,800,3])
plt.figure(figsize=(10,10))
plt.imshow(background.astype('uint8'))
```

Circle to be added in the center of background

```python
back1 = background.copy()

cv2.circle(back1, (400,400), 200, (255,255,255), cv2.FILLED)
plt.figure(figsize=(10,10))
plt.imshow(back1.astype('uint8'))
```

```python
back2 = background.copy()

cv2.rectangle(back2,(50,50),(500,500),(255,255,255),cv2.FILLED)
plt.figure(figsize=(10,10))
plt.imshow(back2.astype('uint8'))
```

```python
end = cv2.bitwise_and(back1, back2)

plt.figure(figsize=(10,10))
plt.imshow(end.astype('uint8'))
```

```python
end = cv2.bitwise_or(back1, back2)

plt.figure(figsize=(10,10))
plt.imshow(end.astype('uint8'))
```

```python
end = cv2.bitwise_xor(back1, back2)

plt.figure(figsize=(10,10))
plt.imshow(end.astype('uint8'))
```

```python
end = cv2.bitwise_not(back1, back2)

plt.figure(figsize=(10,10))
plt.imshow(end.astype('uint8'))
```
