### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'company_sales_data.csv'

data = pd.read_csv(url)

```

```python
df = pd.DataFrame(data)

plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
# Write your code here
x = df['month_number']
y = df['total_units']
plt.figure(figsize=(10, 7))
plt.plot(x, y, '--', color= 'red', linewidth = 3, marker= 'o', markerfacecolor = 'black',  label='total units')

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Company sales data of last year')

plt.grid(False)

plt.legend()
plt.legend(loc='lower right')

```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.
Also, add a grid in the plot (the grid line style should be âââ).

The scatter plot should look like this:


```python
# Write your code here
plt.figure(figsize=(10, 8))
plt.scatter(data["month_number"], data["toothpaste"] , marker='o', linewidth=2, label="Tooth paste Sales data")
plt.grid(True, linestyle='--')
plt.xticks(data["month_number"])
plt.xlabel("Month Number")
plt.title('Toothpaste Sales Data')
plt.ylabel("Number of units Sold")
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
# Write your code here
plt.figure(figsize=(10,8))
width = 0.3
plt.xticks(data["month_number"])
plt.bar(data["month_number"],data["facecream"], width, label="Face Cream sales data")
plt.bar(data["month_number"]+width, data["facewash"], width, label="Face Wash sales data", color="darkorange")
plt.grid(True, linestyle="--")
plt.xlabel("Month Number")
plt.ylabel("Sales unit in number")
plt.legend(loc="upper right")
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.
The histogram should look like this:


```python
# Write your code here
plt.figure(figsize=(10,8))
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(data["total_profit"], label="Profit data", bins=profit_range)
plt.xlabel("profit range in dollar")
plt.ylabel("Actual Profit in dollar")
plt.xticks([150000, 175000, 200000, 225000, 250000, 300000, 350000]); #(; is used to remove extra comments in the output)
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
plt.figure(figsize=(15,8))
total_units = [data["facecream"].sum(),data["facewash"].sum(),data["toothpaste"].sum(),
               data["bathingsoap"].sum(),data["shampoo"].sum(),data["moisturizer"].sum()]
label = ["FaceCream","FaceWash","ToothPaste", "Bathing Soap", "Shampoo", "Moisturizer"]

plt.pie(total_units, labels = label, autopct='%2.1f%%',shadow=True)
plt.title("Sales data")
plt.legend(loc='lower right', bbox_to_anchor=(1.2,0.0))

```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
# Write your code here
fig, axes = plt.subplots(2, sharex=True, figsize=(10,6))
axes[0].plot(data["month_number"],data["bathingsoap"], marker='o', color='k', linewidth=2.5)
axes[0].set_title("Sales data of a Bathingsoap")
axes[1].plot(data["month_number"],data["facewash"], marker='o', color='r', linewidth=2.5)
axes[1].set_title("Sales data of a facewash")
plt.xticks(data["month_number"])
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")

```
