### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'D:/AI-Course/AI-DS-Exercises/AI-DS-Exercises/Chapter 1/8. Matplotlib/company_sales_data.csv'

data = pd.read_csv(url)
#print(data)
df = pd.DataFrame(data)

plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')


```

```python
# Write your code here

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and Facecream Sales Data')

plt.grid(True, linestyle='--')

x_axis = df['month_number']
y_cream = df['facecream']
z_wash = df['facewash']

ax = plt.subplot(111)
ax.bar(x_axis-0.2, y_cream, width=0.4, align='center', label = "Facewash")
ax.bar(x_axis+0.2, z_wash, width=0.4, align='center', label = "Facecream")
plt.legend()

plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here

pie_labels=data.drop(['month_number','total_units','total_profit'],axis = 1).sum(axis=0,skipna=True).keys().tolist()
pie_value=data.drop(['month_number','total_units','total_profit'],axis = 1).sum(axis=0,skipna=True).values.tolist()

fig, ax = plt.subplots(figsize=(10,10))
ax.pie(pie_value, labels=pie_labels)
ax.axis('equal')

ax.legend(loc='lower right', bbox_to_anchor=(1.2,0.0))

plt.show()

```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
# Write your code here
fig, (ax1, ax2) = plt.subplots(2,1,sharex=True,figsize=(12,8))

ax1.plot(data['month_number'],data['bathingsoap'], marker = 'o', c='k', linewidth =4)
ax1.set_title('Sales data of bath soap')

ax2.plot(data['month_number'],data['facewash'], marker = 'o', c='r', linewidth =4)
ax2.set_title('Sales data of facewash')

plt.ylabel('Sales units in number')

fig.show()


```
