```python
url = 'D:\AI-Course\AI-Engineering\AI-Engineering\Chapter 2\Build Week\dataset_halfSecondWindow.csv'
df = pd.read_csv(url)
```

```python
df1 = df[['android.sensor.accelerometer#mean','android.sensor.accelerometer#min','android.sensor.accelerometer#max','android.sensor.accelerometer#std','android.sensor.orientation#mean','android.sensor.orientation#min','android.sensor.orientation#max','android.sensor.orientation#std','android.sensor.linear_acceleration#mean','android.sensor.linear_acceleration#min','android.sensor.linear_acceleration#max','android.sensor.linear_acceleration#std','speed#mean','speed#min','speed#max','speed#std','android.sensor.rotation_vector#mean','android.sensor.rotation_vector#min','android.sensor.rotation_vector#max','android.sensor.rotation_vector#std','android.sensor.game_rotation_vector#mean','android.sensor.game_rotation_vector#min','android.sensor.game_rotation_vector#max','android.sensor.game_rotation_vector#std','sound#mean','sound#min','sound#max','sound#std','android.sensor.gyroscope#mean','android.sensor.gyroscope#min','android.sensor.gyroscope#max','android.sensor.gyroscope#std','android.sensor.gyroscope_uncalibrated#mean','android.sensor.gyroscope_uncalibrated#min','android.sensor.gyroscope_uncalibrated#max','android.sensor.gyroscope_uncalibrated#std']]
X = df1
y = df['target']

```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_i = imputer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_i,y,test_size = 0.2, random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred_lin = svc.predict(X_test)

print("Accuracy of Linear Kernel:",round((accuracy_score(y_test,y_pred_lin)*100),2),"%")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(X_train,y_train)

y_pred_rbf = svc_rbf.predict(X_test)

print("Accuracy of rbf Kernel:",round((accuracy_score(y_test,y_pred_rbf)*100),2),"%")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc_sigmoid = SVC(kernel='sigmoid')
svc_sigmoid.fit(X_train,y_train)

y_pred_sigmoid = svc_sigmoid.predict(X_test)

print("Accuracy of sigmoid Kernel:",round((accuracy_score(y_test,y_pred_sigmoid)*100),2),"%")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc_poly = SVC(kernel='poly')
svc_poly.fit(X_train,y_train)

y_pred_poly = svc_poly.predict(X_test)

print("Accuracy of poly Kernel:",round((accuracy_score(y_test,y_pred_poly)*100),2),"%")
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_dtc = DecisionTreeClassifier(criterion='gini', max_depth=None)
clf_dt = clf_dtc.fit(X_train, y_train)
predictions = clf_dt.predict(X_test)

print("Accuracy of Decision Tree:",round((accuracy_score(y_test, predictions)*100),2),"%")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rfc = RandomForestClassifier(criterion='entropy', max_depth=None, min_samples_split=2, n_estimators=100)
clf_rf = clf_rfc.fit(X_train, y_train)
predictions = clf_rf.predict(X_test)

print("Accuracy of Randomforest Tree:",round((accuracy_score(y_test, predictions)*100),2),"%")
```
